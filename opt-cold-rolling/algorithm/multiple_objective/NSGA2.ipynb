```python
from dataclasses import dataclass
from enum import Enum


# 工序、机器相关参数
class Operation(Enum):
    HR = "hot_rolling",
    AR = "acid_rolling",
    CA = 'continuous_annealing'

@dataclass
class StaticParameters:
    """
    静态参数类

    """
    # 产线顺序：HR -> AR -> CA
    operation_sequence = [Operation.HR, Operation.AR, Operation.CA]

    # 产线之间的运输时间，小时
    transmission_time = {
        Operation.HR: 2.0,
        Operation.AR: 4.0,
        Operation.CA: 1.0 # 最后一站出库
    }

    # 产线不同机组的处理速度，吨/小时
    speed = {
        Operation.HR: [30],
        Operation.AR: [14, 16],
        Operation.CA: [8.0, 12.0, 10.0]
    }
```
